# 1) How have sales and profits changed over time?


```python
total_sales=data['Sales'].sum()
total_profit=data['Profit'].sum()
print('total_sales=',total_sales)
print('total_profit=',total_profit)
```

```python
plt.figure(figsize=(20,5))
sns.set_context('talk')
sns.set_style('dark')
data.groupby(['Order Date'])['Sales'].mean().plot()
data.groupby(['Order Date'])['Profit'].mean().plot()

plt.xlabel('Order Date')
plt.ylabel('usd')

plt.title("Average Sales and Profit over Time Period(2014-2018)")
plt.legend()
plt.show()
```

# 2) what is the most proftuble cateagory? 


```python
category = data.groupby("Category")["Profit"].sum()
plt.figure(figsize=(5, 5)) 
plt.pie(category.values, labels=category.index,autopct="%10.1f%%",explode=(0.04,0.04,0.04))
plt.title("Profit Distribution by Category")
plt.show()
```

# 3) top and bottom cites in sales


```python

plt.figure(figsize=(20, 5))

city_sales = data.groupby('City').agg({'Sales': 'sum', 'Profit': 'sum'})

city_sales['Color'] = city_sales['Profit'].apply(lambda x: 'green' if x > 0 else 'red')


top_cities = city_sales.sort_values(by='Sales', ascending=True).tail(10)


plt.barh(top_cities.index, top_cities['Sales'], color=top_cities['Color'])
plt.legend(labels=['Profit > 0', 'Profit <= 0'])

plt.xlabel('Sales')
plt.ylabel('City')
plt.title('Top 10 Cities by Sales')

plt.show()
```
